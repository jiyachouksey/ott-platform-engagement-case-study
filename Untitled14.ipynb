```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Content Scheduling for Maximizing Engagement.csv")
```

```python
df['Release Date'] = df['Release Date'].astype('category')
df['Genre'] = df['Genre'].fillna('Unknown')
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Set style
sns.set(style="whitegrid", palette="pastel")
plt.figure(figsize=(10, 6))
```

```python
# 1️⃣ Content Type Distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Content Type', data=df)
plt.title("Distribution of Content Types")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
#genre vs viewership
plt.figure(figsize=(10,5))
sns.barplot(x='Genre', y='Viewership', data=df, estimator='mean', errorbar=None)
plt.title("Average Viewership by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
#Day of Week vs Engagement
plt.figure(figsize=(8,5))
sns.boxplot(x='Day of the Week', y='Viewership', data=df)
plt.title("Viewership Distribution by Day of the Week")
plt.show()
```

```python
#Time of Day vs Average Watch Time
plt.figure(figsize=(10,5))
sns.lineplot(x='Time of Day', y='Average Watch Time (mins)', data=df)
plt.title("Average Watch Time Trend Across the Day")
plt.xlabel("Time of Day")
plt.ylabel("Avg Watch Time (mins)")
plt.show()
```

```python
#Viewer Demographics — Gender
plt.figure(figsize=(6,5))
sns.barplot(x='Viewer Gender', y='Viewership', data=df, estimator='mean', errorbar=None)
plt.title("Average Viewership by Gender")
plt.show()
```

```python
#Viewer Age vs Engagement
plt.figure(figsize=(8,5))
sns.boxplot(x='Viewer Age', y='Viewership', data=df)
plt.title("Viewership by Age Group")
plt.show()
```

```python
#Correlation Heatmap (numerical features)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Numerical Features")
plt.show()
```

```python
#  Promotion Effect
plt.figure(figsize=(8,5))
sns.barplot(x='Promotion Data', y='Likes', data=df, estimator='mean', errorbar=None)
plt.title("Impact of Promotions on Likes")
plt.show()
```

```python
#content Duration vs Watch Time
plt.figure(figsize=(8,5))
sns.scatterplot(x='Duration (mins)', y='Average Watch Time (mins)', hue='Genre', data=df)
plt.title("Duration vs Average Watch Time")
plt.show()
```

```python
#Engagement Overview — Likes, Shares, Comments
plt.figure(figsize=(10,6))
sns.histplot(df[['Likes','Shares','Comments']], bins=30, kde=True)
plt.title("Distribution of Engagement Metrics")
plt.xlabel("Engagement Count")
plt.ylabel("Frequency")
plt.show()
```
